## Weather Prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("C:\\Users\\HP\\Downloads\\Climate of England _ Professional Reforms.csv")
data
```

```python
data.rename(columns={'Formatted Date': 'Date'},inplace=True)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
# Handling missing values
data.isnull().sum()
```

```python
# handling missing values
data.ffill(inplace=True)
```

```python
# Preprocess data (example: encoding categorical variables)
data = pd.get_dummies(data, columns=['Summary', 'Precip Type'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Select numerical features to normalize
numerical_features = ['Temperature (C)', 'Apparent Temperature (C)', 'Visibility (km)', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Pressure (millibars)', 'Humidity']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
data[numerical_features] = scaler.fit_transform(data[numerical_features])

print(data.head())
```

```python
# Feature and target selection
features = data.drop(['Date', 'Time', 'Temperature (C)'], axis=1)  # Dropping unnecessary columns
target = data['Temperature (C)']
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, 
                           cv=5, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters and the best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

```python
# Predict and evaluate
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
print(f'Root Mean Squared Error: {rmse}')
```
